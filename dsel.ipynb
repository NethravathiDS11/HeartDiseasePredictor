```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Select two continuous variables for the meshgrid
# Here, let's assume we use 'age' and 'trestbps' (resting blood pressure)
x = data['age']
y = data['trestbps']

# Create the meshgrid
X, Y = np.meshgrid(np.linspace(x.min(), x.max(), 100), np.linspace(y.min(), y.max(), 100))

# Define a function to evaluate on the meshgrid
# For demonstration purposes, let's use a simple linear function: Z = X + Y
# You can replace this with any other function that suits your analysis
Z = X + Y

# Plot the function using a contour plot
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(contour, label='Function value')
plt.title('Contour plot of a linear function (X + Y)')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Select two continuous variables for the quiver plot
x = data['age']
y = data['trestbps']
u = data['chol']
v = data['thalach']

# Creating a subset of the data for clearer visualization
sample_size = 50
indices = np.random.choice(range(len(data)), sample_size, replace=False)
x_sample = x.iloc[indices]
y_sample = y.iloc[indices]
u_sample = u.iloc[indices]
v_sample = v.iloc[indices]

# Create the quiver plot
plt.figure(figsize=(10, 8))
plt.quiver(x_sample, y_sample, u_sample, v_sample, angles='xy', scale_units='xy', scale=1, color='b')
plt.title('Quiver Plot of Sample Data')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_csv('/home/nethranaidu/Downloads/heartdataset.csv')

# Ensure 'target' column exists and separate features and target
X = data.drop('target', axis=1)
y = data['target']

# Verify shapes
print(X.shape)
print(y.shape)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the training data, transform the testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the k-NN classifier with k=5 (or any other k value)
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the model on the scaled training data
knn.fit(X_train_scaled, y_train)

# Predict on the testing data
y_pred = knn.predict(X_test_scaled)

# Print the accuracy score
print(f'Accuracy: {accuracy_score(y_test, y_pred):.2f}')

# Print the classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Print the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

# Plot the confusion matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```
